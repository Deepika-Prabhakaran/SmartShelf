```python
### 20 method
import cv2
from pyzbar.pyzbar import decode
from datetime import datetime
# Import your database library (e.g., Firebase)

cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Set width
cap.set(4, 480)  # Set height

while True:
    success, img = cap.read()  # Capture frame from webcam
    if not success:
        break  # If frame is not captured correctly, break the loop

    for barcode in decode(img):
        mydata = barcode.data.decode('utf-8')
        l = []
        for i in mydata.split(','):
            l.append(i.split(':'))
        print(l)
        
        # Get the current date and time
        now = datetime.now()

        # Format the date and time
        formatted_date_time = now.strftime("%Y-%m-%d %H:%M:%S")

        # Assuming `db` is your database reference
        doc1 = db.collection('srn')
        docs = doc1.get()
        for doc in docs:
            if doc.exists and list((doc.to_dict()).values())[0] == '0':
                doc_ref1 = db.collection('srn').document(doc.id)
                doc_ref1.set({
                    l[-1][1]: l[0][1]
                })
                doc_ref = db.collection('warehouse').document()
                doc_ref.set({
                    'product_name': l[0][1],
                    'Quantity': l[-2][1],
                    'type': l[-3][1],
                    'Manf_name': l[1][1],
                    'product_id': l[-1][1],
                    'price': l[2][1],
                    'srn': doc.id,
                    'time': str(formatted_date_time)
                })
                break

    # Show the captured frame
    cv2.imshow("original", img)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
################################################################################
# stock request by the inventory from the warehouse

# Initialize video capture
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

while True:

    success, img = cap.read()
    if not success:
        break
    l=[]
    for barcode in decode(img):
        mydata = barcode.data.decode('utf-8')
        for i in mydata.split(','):
            l.append(i.split(':'))
        print(l)
        product_name=l[0][-1]
        quantity_required = int(l[-2][-1])
        manf_name=l[1][-1]
        type=l[3][-1]
        sell_price=l[2][-1]
        product_id=l[-1][-1]
        
        # Search the warehouse database by product name
        warehouse_ref = db.collection('warehouse')
        query = warehouse_ref.where('product_name', '==', product_name).stream()

        found = False
        for doc in query:
            product_data = doc.to_dict()
            print(product_data)
            available_quantity = int(product_data['quantity_in_stock'])
            manf_name = product_data['manf_name']
            sell_price = float(product_data['price'])
            product_type = product_data['type']

            if available_quantity > 0:
                # Determine the quantity to use (either the requested amount or the available amount)
                quantity_to_use = min(quantity_required, available_quantity)
                
                # Update quantity in the warehouse
                new_quantity = available_quantity - quantity_to_use
                
                if new_quantity > 0:
                    warehouse_ref.document(doc.id).update({'quantity_in_stock': new_quantity})
                else:
                    warehouse_ref.document(doc.id).delete()  # Remove product if quantity becomes zero
                
                # Add product to the inventory (user's order)
                inventory_ref = db.collection('inventory').document()
                inventory_ref.set({
                    'manf_name': manf_name,
                    'product_id': doc.id,  # Firebase unique ID
                    'quantity': quantity_to_use,
                    'product_name': product_name,
                    'sell_price': sell_price,
                    'type': product_type
                })
                
                print(f"Product {product_name} with quantity {quantity_to_use} added to the inventory.")
                print(f"Available stock for {product_name} was {available_quantity}.")
                found = True
                break
        
        if not found:
            print(f"Product {product_name} not found in the warehouse or no stock available.")

cap.release()
cv2.destroyAllWindows()
```
